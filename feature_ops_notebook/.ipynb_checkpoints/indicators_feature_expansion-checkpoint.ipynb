```python
from ta import add_all_ta_features
import pandas as pd
```

```python
df_aapl = pd.read_csv('./data/eod/AAPL.csv')
df_aapl
```

```python
df_aapl_added = add_all_ta_features(
    df_aapl, open="Open", high="High", low="Low", close="Close", volume="Volume", fillna=True)
df_aapl_added = df_aapl_added[20:]
```

```python
df_aapl_added.reset_index(drop=True)
```

```python
symbols = ['TSLA', 'AAPL', 'AMZN', 'GOOG','NFLX', 'FB']
for symbol in symbols:
    df = pd.read_csv('./data/eod/{}.csv'.format(symbol))
    df = add_all_ta_features(
        df, open="Open", high="High", low="Low", close="Close", volume="Volume", fillna=True)
    df = df[20:]
    df = df.reset_index(drop=True)
    df.to_csv('./data/eod/{}_features.csv'.format(symbol), encoding='utf-8')
```

```python
symbols = ['TSLA', 'AAPL', 'AMZN', 'GOOG','NFLX', 'FB']
periods = [['2020-04-29','2020-10-26'],
          ['2020-10-27','2021-04-23'],
          ['2021-04-24','2021-10-21'],
          ['2021-10-22','2022-04-20'],
          ['2022-04-21','2022-04-29']]
for symbol in symbols:
    df_list = []
    for period in periods:
        params = {
            'api_token': 'v4e5qEOulw9UJn3WZWA1lvLV9py2R4hxtOHHrK8n',
            'symbols': symbol,
            'date_from': period[0],
            'date_to': period[1],
            'interval': 'hour',
            'language': 'en'
        }

        api_result = requests.get('https://api.stockdata.org/v1/data/intraday?', params)

        api_response = api_result.json()

        df = pd.json_normalize(api_response['data'])
        
        df_list.insert(0,df)
    df_all = pd.concat(df_list, ignore_index=True)
    df_all.to_csv('./data/hourly/{}.csv'.format(symbol), encoding='utf-8')
```
